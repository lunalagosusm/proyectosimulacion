A continuacion, se presenta el codigo python encargado de procesar los archivos de texto resultantes de la simulacion del codigo en C:

```python
# Librerias necesarias para realizar graficas
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import numpy as np
```

```python
lista1 = [] 
lista2 = []
archivo1 = open("resultado1.txt","r") # Este archivo almacena la prob. bloqueo total de cada simulacion
archivo2 = open("resultado2.txt","r") # Este archivo añmacena la prob. de bloqueo cada 100 muestars por simulacion

# se realiza la separacion de datos de ambos archivos, almacenandose todos en una lista de strings
for linea in archivo1.readlines():
    lista1.append(linea.split(":"))
for linea in archivo2.readlines():
    lista2.append(linea.split(":"))
    
# Posteriormente se realiza un proceso de formato a ambas listas, convirtiendo los datos de string a float
datos1=[[],[],[],[],[],[],[],[]] #Prob. Bloqueo total
datos2=[[],[],[],[],[],[],[],[]] #Prob. Bloqueo cada 100 muestras

i=0
for i in range(8):
    for element in lista1[i]:
        element = float(element)
        datos1[i].append(element)

j=0
for j in range(8):
    lista2[j].remove('\n')
    for element in lista2[j]:
        element = float(element)
        datos2[j].append(element)
```
