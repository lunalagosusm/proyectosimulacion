```python
import matplotlib.pyplot as plt
import numpy as np
import math

paleta=["#073059","#2866ab","#5fbdc5","#d8d95c","#FF0000","#FFFFFF"]
```

```python
graphs = [line.rstrip('\n') for line in open('Datos_a_graficar.txt')]
graph1 = graphs[0].split(":")
graph1.pop(-1)
graph2 = graphs[1].split(":")
graph2.pop(-1)
graph3 = graphs[2].split(":")
graph3.pop(-1)
graph4 = graphs[3].split(":")
graph4.pop(-1)

x1=[]
y1=[]
for linea in graph1:
    var=linea.strip().split("	")
    x1.append(float(var[0]))
    y1.append(float(var[1]))
    
x2=[]
y2=[]
for linea in graph2:
    var=linea.strip().split("	")
    x2.append(float(var[0]))
    y2.append(float(var[1]))
    
x3=[]
y3=[]
for linea in graph3:
    var=linea.strip().split("	")
    x3.append(float(var[0]))
    y3.append(float(var[1]))
    
x4=[]
y4=[]
for linea in graph4:
    var=linea.strip().split("	")
    x4.append(float(var[0]))
    y4.append(float(var[1]))
```

```python
plt.figure(1)
plt.axes(facecolor=paleta[5])
plt.grid()
plt.title("\nSIMULACION ArpaNet\n" + r"P. de Bloqueo V/S Nmero de Canales",fontsize=25)
plt.xlabel('Nmero de Canales',fontsize=18)
plt.ylabel(r'Probabilidad de Bloqueo',fontsize=18)
plt.plot(x1,y1,color=paleta[4],label=r"Probabilidad de Bloqueo Sin Cortes")
plt.plot(x2,y2,color=paleta[3],label=r"Probabilidad de Bloqueo Corte Enlace 2")
plt.plot(x3,y3,color=paleta[2],label=r"Probabilidad de Bloqueo Corte Enlace 18")
plt.plot(x4,y4,color=paleta[1],label=r"Probabilidad de Bloqueo Corte Enlace 16")
plt.legend(loc = 1,fontsize=15)
plt.subplots_adjust(top=1.5,right=2)
axes = plt.gca()
axes.set_xlim([10,26])
axes.set_ylim([0,0.01])
plt.show()
```
